```python
# BAI 3
import math
n = int(input("Nhap n = "))
a = abs(n)
if a == 0 :
    count = 0
    total = 1
else :
    count = 0
    total = 0
    while a > 0 :
        digit = a % 10
        total += digit 
        count += 1
        a //= 10
    print("So chu so la ",count)
    print("Tong cac chu so la ",total)

#kiem tra so nguyen to 
if n < 2 :
    print (n, "khong phai la so nguyen to")
else :
    prime = True 
    for i in range(2, int(math.sqrt(n) + 1)) :
        if n % i == 0 :
            prime = False
            break
    if prime :
        print(n,"la so nguyen to")
    else :
        print(n,"khong la so nguyen to")
```

```python
# BAI 4
money = int(input("Nhap so tien : "))
beer = money // 28
total = beer
shell = beer
while shell >= 3 :
    new_beer = shell // 3
    total += new_beer
    shell = shell % 3 + new_beer
print("Tong so bia co the uong:",total)
```
